```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

df_folds = df[['path']].copy()
# 일단 박스 하나당 1이라는 index를 주어줌 (박스별로 나누기 위함)
df_folds.loc[:, 'bbox_count'] = 1

df_folds
```

```python
df_folds = df_folds.groupby('path').count()

df_folds
```

```python
# df_folds에 label 축 생성
df_folds.loc[:, 'label'] = df[['path', 'label']].groupby('path').min()['label']

# df_folds['label'].unique()
df_folds
```

```python
df_folds.bbox_count.describe()
```

```python
 df_folds['bbox_count'].apply(lambda x: f'_{x // 3}').values.astype(str)
```

```python
# df_folds에 stratify_group 축 생성

# bbox_count는 한 이미지 내에 있는 전체 box의 갯수
# 만약 한 이미지내에 box의 갯수가 지나치게 많은 경우 라벨링을 1_0, 1_1, 1_2 ... 로 이중 라벨링을 매겨서 세분화 한 뒤 학습
# 실제로 GWD에서는 한 이미지에 박스가 매우 많았음
df_folds.loc[:, 'stratify_group'] = np.char.add(
    df_folds['label'].values.astype(str),
    df_folds['bbox_count'].apply(lambda x: f'_{x // 3}').values.astype(str)
)

# df_folds['stratify_group'].unique()

df_folds
```

```python
# df_folds에 fold 축 생성
df_folds.loc[:, 'fold'] = 0

df_folds
```

```python
folds = [(idxT,idxV) for i,(idxT,idxV) in enumerate(skf.split(np.arange(df_folds.shape[0]), df_folds['stratify_group']))]
for i in range(5):
    (idxT,idxV) = folds[i]
    
    print(idxT)
```
